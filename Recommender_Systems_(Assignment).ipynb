```python
books = pd.read_excel('books.xlsx')
books
```

```python
# CHANGING THE COLUMN NAMES FOR OUR CONVINIENCE 

BOOKS1 = books.rename({'User.ID':'UID','Book.Title':'Title','Book.Rating':'Rating'}, axis=1)
BOOKS1
```

```python
BOOKS1.shape
```

```python
len(BOOKS1.Title.unique())
```

```python
len(BOOKS1.UID.unique())
```

```python
# Sorting the UID
```

```python
BOOKS1.sort_values('UID')
```

```python
BOOKS = BOOKS1.drop_duplicates('UID')
```

```python
BOOKS
```

```python
user_books_df = BOOKS.pivot(index='UID', columns='Title', values='Rating').reset_index(drop=True)
```

```python
user_books_df
```

```python
# Input those NAN's with 0 value
```

```python
user_books_df.fillna(0,inplace=True)
```

```python
user_books_df
```

```python
# Calculating Cosine Similarity between users
```

```python
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation

user_sim = 1 - pairwise_distances(user_books_df.values,metric='cosine')

user_sim
```

```python
# Store the result in a dataframe
```

```python
user_sim_df = pd.DataFrame(user_sim)
```

```python
user_sim_df
```

```python
user_sim_df.index  =  BOOKS.UID.unique()
user_sim_df.columns = BOOKS.UID.unique()
```
